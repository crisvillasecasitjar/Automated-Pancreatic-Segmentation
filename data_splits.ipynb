```python
final_scan_path = []
final_scans_list, final_labels_list = [], []

# Directorio donde están las etiquetas
labels_dir = "/data/PANORAMA/cvillaseca/panorama_labels"  # Reemplaza con la ruta correcta

for n in os.listdir(scans_path):
    scan_path = osp.join(scans_path, n)
    file_name = osp.basename(scan_path)

    # Verifica si la etiqueta correspondiente está en la lista filtered_labels
    study_id = file_name[:-12]
    manual_label = osp.join(labels_dir, 'manual_labels', study_id + '.nii.gz')
    automatic_label = osp.join(labels_dir, 'automatic_labels', study_id + '.nii.gz')

    if manual_label in filtered_labels:
        final_labels_list.append(manual_label)
        final_scans_list.append(scan_path)
    elif automatic_label in filtered_labels:
        final_labels_list.append(automatic_label)
        final_scans_list.append(scan_path)
    else:
        print('Useless label:', scan_path)

print(f'Total useful scans: {len(final_scans_list)}')
print(f'Total useful labels: {len(final_labels_list)}')
```
